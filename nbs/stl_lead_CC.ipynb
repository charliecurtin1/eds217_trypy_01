```python
import pandas as pd

url = 'https://raw.githubusercontent.com/'\
'allisonhorst/EDS_221_programming-essentials/'\
'main/activities/stl_blood_lead.csv'

stl_blood_lead = pd.read_csv(url)
```

```python
stl_blood_lead['prop_white'] = (stl_blood_lead['white'] / stl_blood_lead['totalPop']) * 100

stl_blood_lead.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
y = stl_blood_lead['pctElevated']
x = stl_blood_lead['prop_white']
plt.scatter(x, y, alpha=.5, s=20)
plt.title('Blood Lead vs Proportion White')
plt.ylabel('Blood Lead')
plt.xlabel('Proportion White')
```

```python
plt.hist(y, alpha=.9, bins=20, color='pink', edgecolor='black')
plt.xlabel('Percentage Elevated Blood Lead')
plt.ylabel('Frequency')
plt.title('Histogram of Percentage Elevated Blood Lead', fontweight='bold', fontsize=15)
fig = plt.gcf()
fig.set_facecolor('#FFFDD0')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0, 100)
plt.savefig('histogram.png', dpi=300)
```
